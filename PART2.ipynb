```python
system_instruction = f"""
You are "BiteBot", the helpful, friendly, and efficient AI assistant for the "Campus Bites Canteen" on campus.

Your job is strictly to answer questions about the canteen using ONLY the provided Knowledge Base below.

CRITICAL RULES:
1. You must answer questions grounded ONLY in the provided Knowledge Base.
2. If a user asks about something NOT mentioned in the Knowledge Base (e.g., homework help, general knowledge, weather, or items not on the menu), you must politely refuse to answer. Use phrases like: "I am sorry, but I can only answer questions related to the Campus Bites Canteen."
3. Keep your answers concise, direct, and factual. Do not make up prices, items, or rules.

KNOWLEDGE BASE:
{knowledge_base}
"""
```

```python
model = genai.GenerativeModel(
    model_name="gemini-1.5-flash",
    system_instruction=system_instruction,
    generation_config={"temperature": 0.25} # Keeps the bot strictly factual
)
```

```python
def start_canteen_bot():
    # start_chat() handles the conversation history/memory automatically
    chat = model.start_chat(history=[])

    print("="*60)
    print("🍔 Welcome to Campus Bites Canteen Chatbot (BiteBot)! 🍟")
    print("Ask me anything about our timings, menu, location, or rules.")
    print("Type 'quit' or 'exit' to end the conversation.")
    print("="*60)

    while True:
        user_input = input("\nYou: ").strip()

        # Check for exit commands gracefully
        if user_input.lower() in ['quit', 'exit']:
            print("\nBiteBot: Thank you for visiting Campus Bites Canteen! Have a great day ahead! 👋")
            break

        if not user_input:
            continue

        # Send message to the multi-turn chat session
        try:
            response = chat.send_message(user_input)
            print(f"\nBiteBot: {response.text}")
        except Exception as e:
            print(f"\nAn error occurred: {e}")

# Run the chatbot program
if __name__ == "__main__":
    start_canteen_bot()
```
